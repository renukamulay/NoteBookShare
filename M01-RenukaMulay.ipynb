Import a Csv File and prints first N Rows's First and Third Column.

```python
import json as json
import numpy as np
import pandas as pd
from IPython.display import display

def PrintTitleAndTotalPagesOfBooks(fileName = ""):
    if fileName == "":
        return
    """Parses few key columns from a csv file and loads the data in Pandas dataFrame"""
    data = []
    with open(fileName, newline='') as jsonDoc:  
        for row in json.load(jsonDoc):
            datarow = []
            isbn = None
            pageCount = None
            title = None
            desc_len = None
            has_word_data = None;
            if "isbn" in row:
                isbn = row["isbn"]
            if "pageCount" in row:
                pageCount = row["pageCount"]
            if "title" in row:
                title = row["title"]
            if "longDescription" in row:
                desc_len = len(row["longDescription"].split(" "))
                has_word_data = row["longDescription"].find("data") != -1
            
            datarow.append(len(row["authors"]))
            datarow.append(isbn)
            datarow.append(pageCount)
            datarow.append(title)
            datarow.append(desc_len)
            datarow.append(has_word_data)
            data.append(datarow)
    df = pd.DataFrame(np.array(data), 
                      columns=['Authors_Length', 'ISBN', 'PageCount', 'Title', 'Desc_Length', 'Has_Data'])
    display(df.head(20))

            
PrintTitleAndTotalPagesOfBooks('data/books.json')
```
