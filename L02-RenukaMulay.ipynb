Import a Csv File and prints first N Rows's First and Third Column.

```python
import json as json

def PrintTitleAndTotalPagesOfBooks(fileName = "", rowCount = 5):
    if fileName == "":
        return
    """Returns Book Title and Total Pages of first N books"""
    with open(fileName, newline='') as jsonDoc:  
        for row in json.load(jsonDoc):
            print(f'Title = {row["authors"]} Length = {len(row["authors"])}')

PrintTitleAndTotalPagesOfBooks('data/books.json', 2)
```
